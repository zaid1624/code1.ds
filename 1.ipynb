```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated stock price data
np.random.seed(42)
num_days = 200
days = np.arange(num_days)
stock_prices = 100 + 2 * days + np.random.normal(scale=10, size=num_days)

# Split data into features (days) and labels (prices)
X = days.reshape(-1, 1)
y = stock_prices

# Build a linear regression model
model = np.polyfit(X.flatten(), y, 1)
predict = np.poly1d(model)

# Predict future stock prices
future_days = np.arange(num_days, num_days + 10)
future_prices = predict(future_days)

# Plot original data and predicted future prices
plt.scatter(days, stock_prices, label='Original Data')
plt.plot(future_days, future_prices, label='Predicted Future Prices', color='orange')
plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
